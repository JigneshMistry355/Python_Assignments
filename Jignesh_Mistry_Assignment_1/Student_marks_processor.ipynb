```python
import pandas as pd
import numpy as np
```

# Creating data

```python
df = pd.DataFrame(
    {
        "registration_number": [10001, 10002, 10003, 10004, 10005],
        "Exam_marks": [57, 60, 71, 65, 69],
        "Coursework_marks": [21, 19, 20, 17, 23]
    }
)
df_csv = df.to_csv("Student_data.csv", index= False)
```

# 1. Reading Data from external file

```python
student_data = pd.read_csv("Student_data.csv")
student_data.head()
```

# 2. Compute overall marks

```python
student_data["Overall"] = student_data["Exam_marks"] + student_data["Coursework_marks"]
student_data
```

# 3. Assign Grades

```python
def get_grades(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    elif marks >= 60:
        return "D"
    elif marks >= 50:
        return "E"
    else: 
        return "F"
        
student_data["Grade"] = student_data["Overall"].apply(get_grades)
student_data
```

# 4. Sort by overall marks

```python
student_data = student_data.sort_values(by=["Overall"], ascending=False)
student_data
```

# 5. Write data to external file

```python
student_sorted_data = student_data.to_csv("student_sorted_data.csv", index=False)
```
